```python
import pandas as pd
```

```python
df = pd.read_csv('runs.csv').drop(columns=['Unnamed: 0', 'run_id', 'experiment_id', 'tags.mlflow.source.name', 'tags.mlflow.source.git.commit',
                                           'status', 'artifact_uri', 'start_time', 'end_time', 'tags.mlflow.source.type', 'tags.mlflow.user'])
df
```

```python
for file in ['TSPC.csv', 'TSPD.csv']:
    for greedy in [True, False]:
        for type_of_change in ['edge', 'node']:
            
            df_final = df[df['params.lab'] == 3]            
            df_final = df_final[df_final['params.file_name'] == file] 
            df_final = df_final[df_final['params.greedy'] == greedy]
            df_final = df_final[df_final['params.type_of_change'] == type_of_change]
            df_final = df_final[df_final['params.start_node_idx'].notna()]
            print(file, greedy, type_of_change)
            display(df_final[['metrics.cost', 'metrics.time', ]].describe().loc[['mean', 'min', 'max']])
```
